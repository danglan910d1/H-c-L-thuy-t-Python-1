```python
std = {"hs1":"Lan", "hs2":"Hà", "hs3":"Đặng", "hs4":"Trần"}
print(std, type(std))
```

```python
# Đếm tổng số phần tử trong Dictionary
len(std)
```

```python
print(std['hs1'], std['hs3'])
print(std['hs2'], std['hs4'])
```

```python
# Xoá một phần tử ở vị trí index
# del std['hs2']
```

```python
# Lấy ra khoá của Dictionary
std.keys()
```

```python
# Lấy ra giá trị của Dictionary
std.values()
```

```python
# Hiển thị ra đầy đủ key và values của Dictionary
i = std.items()
for item in i :
  print(item[0])
  print(item[1])
  print(item)
```

```python
diem = dict(A = 10, B = 5, C = 7)
diem, type(diem)
```

```python
# Chuyển dict sang list thì kết quả trả ra chỉ giữ lại key
list(diem)
```

```python
# Dùng hàm tối ưu hóa để xây dựng dict
result = {x:x**2 for x in range(10)}
result 
```
