```python
# Bài 4: Xây dựng hàm chạy tử 1 đến 100 và chỉ in ra khi là số nguyên tố. Xuất ra thành các cột đều nhau
def songuyento(n) :
    if n <= 1 :
        return False
    elif n == 2 :
        return True
    else :
        for i in range(2, n) :
            if n % i == 0 :
                return False
            return True
dem = 0
for i in range(1, 1000) :
    if songuyento(i) :
        dem = dem + 1
        print("{:>3}".format(str(i)), end = " ")
        if dem % 10 == 0 :
            print("")

```
